### 중첩 for문을 이용한 구구단 프로그램

```python
for i in range(2, 10):
    print('{}단'.format(i))
    for j in range(1, 10):
        print(i, '*', j, '=', i*j)
    print()
```

### for문을 이용한 369게임

```python
final = int(input('얼마까지 369할까? '))

for i in range(1, final+1):
    cnt = 0
    for j in str(i):
        int_j = int(j)
        if int_j % 3 == 0 and int_j != 0:
            cnt += 1
    if cnt != 0:
        print(i, '->', 'clap!! '*cnt)
    if cnt == 0:
        print(i, '->', i)

```
